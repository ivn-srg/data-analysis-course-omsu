# Библиотеки для получения и обработки данных


```python
# 1. Работа с модулем math
import math

print(math.sqrt(16))      # 4.0 - квадратный корень
print(math.factorial(5))  # 120 - факториал числа
print(math.pi)            # 3.141592653589793 - значение π
print(math.sin(math.radians(30)))  # 0.5 - синус 30 градусов
```

```python
# 2. Работа с модулем numpy
import numpy as np

# Создание массива numpy
arr = np.array([1, 2, 3, 4, 5])
print(arr * 2)  # [ 2  4  6  8 10] - умножение массива на число
```

```python
# Создание массива с нулями, единицами и случайными числами
zeros = np.zeros((2, 3))
ones = np.ones((3, 3))
random_arr = np.random.rand(3, 3)

print(zeros)
print(ones)
print(random_arr)
```

```python
# Операции над массивами
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
sum_arr = arr1 + arr2  # [5 7 9]
double_product = (arr1 + arr2) * 2

print(sum_arr)
print(double_product)
```

```python
# Матрицы и их операции
matrix = np.array([[1, 2], [3, 4]])
transpose = matrix.T  # Транспонирование
inverse = np.linalg.inv(matrix)  # Обратная матрица

print(transpose)
print(inverse)
```

```python
# 3. Работа с модулем pandas
import pandas

# Создание DataFrame
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "City": ["New York", "Los Angeles", "Chicago"]
}

df = pandas.DataFrame(data)

df
```

## 📖 Кодбук для данных о покупателях

| Переменная               | Описание                                        | Тип данных     | Пример значений                     |
|--------------------------|------------------------------------------------|---------------|--------------------------------------|
| `Customer ID`           | Уникальный идентификатор покупателя            | Целочисленный  | 10245, 58963, 74125                 |
| `Purchase Date`         | Дата покупки                                   | Дата          | 2023-05-12, 2024-01-30              |
| `Product Category`      | Категория купленного товара                    | Строка        | "Электроника", "Одежда", "Книги"    |
| `Product Price`         | Цена одного товара                            | Число (float) | 1499.99, 250.50, 79.99              |
| `Quantity`              | Количество купленных товаров                   | Целочисленный  | 1, 3, 5                             |
| `Total Purchase Amount` | Общая сумма покупки                           | Число (float) | 2999.98, 1250.75                    |
| `Payment Method`        | Способ оплаты                                  | Строка        | "Карта", "Наличные", "PayPal"       |
| `Customer Age`          | Возраст покупателя                            | Целочисленный  | 25, 40, 63                          |
| `Returns`               | Количество возвратов                           | Целочисленный  | 0, 1, 2                             |
| `Customer Name`         | Имя покупателя                                | Строка        | "Иван Петров", "Анна Смирнова"      |
| `Age`                   | Дублирующий возраст (аналог `Customer Age`)    | Целочисленный  | 25, 40, 63                          |
| `Gender`                | Пол покупателя                                | Строка        | "Мужчина", "Женщина"                |
| `Churn`                 | Покупатель ушел (1) или остался (0)           | Булевый (0/1) | 0, 1                                |

📌 **Примечание:**  
- Поле `Age` дублирует `Customer Age`, возможно, стоит оставить только одно.  
- `Churn = 1` означает, что покупатель перестал делать покупки.  
- `Returns > 0` означает, что покупатель возвращал хотя бы один товар.
